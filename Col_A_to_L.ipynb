```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt

df = pd.read_csv('dataset/initial.csv')
columns_list = df.columns.values.tolist()
print(columns_list)

first = df[columns_list[0]]
second = df[columns_list[1]]
third = df[columns_list[2]]
fourth = df[columns_list[3]]
fifth = df[columns_list[4]]
sixth = df[columns_list[5]]
seventh = df[columns_list[6]]
eighth = df[columns_list[7]]
ninth = df[columns_list[8]]
tenth = df[columns_list[9]]
eleventh = df[columns_list[10]]
twelfth = df[columns_list[11]]

```

```python
for i in range(12):
    print(columns_list[i])
    print("Number of missing value : "+str(df[columns_list[i]].isnull().sum()))
    print("Percentage of missing values : " + str(df[columns_list[i]].isnull().sum()/len(df[columns_list[i]])*100))
    print("Number of unique value : "+str(df[columns_list[i]].nunique())+"\n\n")
```

```python
numerical_cols = ['time_in_hospital']
for col in numerical_cols:
    print(col)
    toAnalyse = df[col].dropna()
    print(toAnalyse)
    print("Mean of the values: "+str(toAnalyse.mean()))
    print("Median of the values: "+str(toAnalyse.median()))
    print("Mode of the values: "+str(toAnalyse.mode()))
    print("Standard deviation of the values: "+str(toAnalyse.std()))
    print("Variance of the values: "+str(toAnalyse.var()))
    print("Skewness of the values: "+str(toAnalyse.skew()))
    print("Kurtosis of the values: "+str(toAnalyse.kurtosis()))
    print("Minimum value: "+str(toAnalyse.min()))
    print("Maximum value: "+str(toAnalyse.max()))
    print("Range of the values: "+str(toAnalyse.max()-df[col].min()))
    print("25th percentile: "+str(toAnalyse.quantile(0.25)))
    print("50th percentile: "+str(toAnalyse.quantile(0.50)))
    print("75th percentile: "+str(toAnalyse.quantile(0.75)))
    print("Interquartile range: "+str(toAnalyse.quantile(0.75)-df[col].quantile(0.25))+"\n\n")
    plt.title(col)
    plt.boxplot(toAnalyse.tolist())

```

```python
pie_chart_cols = ['race', 'gender', 'payer_code', 'medical_specialty']
for col in pie_chart_cols:
    print(col)
    toAnalyse = df[col].dropna()
    plt.title(col)
    plt.pie(toAnalyse.value_counts(), labels=toAnalyse.unique(), autopct='%1.1f%%', shadow=True, startangle=140)
    plt.axis('equal')
    plt.show()

```
